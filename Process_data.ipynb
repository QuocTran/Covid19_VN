```python
def normalize_tinh_thanh (tinh_thanh):
    if tinh_thanh in ['ĐỒNG NAI', 'ĐỒNG NAO', 'Đồng Nai']:
        return 'ĐỒNG NAI'
    elif tinh_thanh in ['BÀ RỊA VŨNG TÀU', 'Bà Rịa-V.Tàu', 'VŨNG TÀU']:
        return 'BÀ RỊA VŨNG TÀU'
    elif tinh_thanh in ['ĐÁK LẮK', 'ĐĂKS LẮK', 'ĐẮC LẮC', 'ĐẮK LẮK', 'Đắk Lắk']:
        return 'ĐẮK LẮK'
    elif tinh_thanh in ['HUẾ', 'TT- Huế']:
        return 'TT- Huế'
    elif tinh_thanh in ['KIEN GIANG', 'KIÊN GIANG']:
        return 'KIÊN GIANG'
    elif tinh_thanh in ['VINH LONG', 'VĨNH LONG']:
        return 'VĨNH LONG'
    else:
        return tinh_thanh.upper()
```
