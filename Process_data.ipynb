```python
import pandas as pd
import numpy as np
```

```python
def get_data(sheet_url):
    url_1 = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
    print(url_1)
    data = pd.read_csv(url_1).set_index('DateTaken')
    data.index = pd.to_datetime(data.index)
    return data.sort_index()
```

```python
data = get_data("https://docs.google.com/spreadsheets/d/12hsnTzHVuylHWTn2fl6TZXgk2JB3UpShjxApPqjCeQ4/edit#gid=488865780")
```
